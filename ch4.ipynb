```julia
function polygon(t, n, len)
    for i in 1:n
        angle = 360/n
        forward(t, len)
        turn(t, -angle)
    end 
end 
```

```julia
üê¢ = Turtle()
@svg begin
    polygon(üê¢, 8, 70)
end
```

```julia
function circle(t, r)
    circumference = 2*pi*r
    n = trunc(circumference/3)+3
    len = circumference / n 
    polygon(t, n, len)
end
```

```julia
üê¢ = Turtle()
@svg begin
    circle(üê¢, 100)
end
```
